### STROKE STATISTICS

```python
df[df["stroke"]==1].describe()
```

#### 1. People having age greater than 67 are more likely to suffer from stroke.

```python
df_stk=df[df["stroke"]==1]
df_stk.head()
```

### STROKE AND AGE 

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df["age"],shade=True)
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Distribution of Age")
```

#### This graph shows the distribution of age in the dataset.

```python
stroke=df[df["stroke"]==1].age
no_stroke=df[df["stroke"]==0].age
fig=ff.create_distplot([stroke,no_stroke],["Stroke","No Stroke"],curve_type="normal")
fig.update_layout(title="Distribution of Age" , xaxis_title="age")
```

#### People having age more than 50 are more likely to suffer from stroke. 

### STROKE AND GENDER

```python
df["gender"].value_counts().plot(kind="bar")
plt.title("distribution of gender")
```

```python
sns.violinplot(x="gender",y="stroke",data=df)
```

#### Violin plot is used for visualising probability distribution

```python
df.groupby("gender").stroke.mean()
```

##### male has 5.1% chances of having stroke and women have 4.7% chance.

### STROKE AND HYPERTENSION

```python
df["hypertension"].value_counts().plot(kind= "bar")
```

```python
sns.catplot(x="hypertension",hue="stroke",kind="count",data=df)
```

```python
df.groupby("hypertension").stroke.mean()
```

#### 13.25% people having hypertension have stroke and 3.96% people not having hypertension suffer from stroke

### STROKE AND HEART DISEASE

```python
df.groupby("heart_disease").stroke.mean()
```

```python
sns.catplot(x="hypertension",hue="stroke",kind="count",data=df)
```

#### People having heart disease tend to have stroke by 17.02% and people having no heart disease have a 4.17% of having a stroke

### STROKE AND AVERAGE GLUCOSE LEVEL

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df_stk["avg_glucose_level"],shade=True)
plt.xlabel("avg_glucose_level")
plt.ylabel("Density")
plt.title("Distribution of Avg glucose level")
```

```python
stroke=df[df["stroke"]==1].avg_glucose_level
no_stroke=df[df["stroke"]==0].avg_glucose_level
fig=ff.create_distplot([stroke,no_stroke],["Stroke","No Stroke"],curve_type="normal")
fig.update_layout(title="Distribution of Avg glucose Level" , xaxis_title="avg_glucose_level")
```

#### People having avg glucose level less than 150 and people having avg glucose level greater than 200 are more likely to suffer from stroke.

### STROKE AND BMI


```python
plt.figure(figsize=(10,5))
sns.kdeplot(df["bmi"],shade=True)
plt.xlabel("bmi")
plt.ylabel("Density")
plt.title("Distribution of BMI")
```

```python
stroke=df[df["stroke"]==1].bmi
no_stroke=df[df["stroke"]==0].bmi
fig=ff.create_distplot([stroke,no_stroke],["Stroke","No Stroke"],curve_type="normal",bin_size=2)
fig.update_layout(title="Distribution of BMI" , xaxis_title="BMI")
```

#### People having bmi greater than 25 have more chances of stroke
